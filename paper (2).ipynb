# time

```python
revised_time=[]
for i in range(len(time_data)):
    revised_time.append(round((time_data[i]-65378)/30))

print(revised_time)
```

# test

```python
lat_data[50],lon_data[51] #한국
```

```python
test=[]
x=len(revised_time)
for i in range(x):
    test.append(precip_data[i][50][51])

plt.figure(figsize=(10,5))
plt.plot(revised_time,test)
plt.title('Korea precipitaion')
plt.xlabel('time')
plt.ylabel('precipitation');
```

```python
fft=np.fft.fft(test)/len(test)
fft_magnitude = abs(fft)
fft_revised=fft_magnitude[1:-1] #[0]데이터를 넣으면 0hz에서 치솟는 데이터가 나와서 [1:-1]로 설정
plt.plot(fft_revised)
```

```python
#fft test1

fshift=np.fft.fftshift(fft_revised)
f=np.linspace(-365/2,365/2,len(fft_revised)) #x축은 어떻게?
plt.plot(f,fshift)
```

```python
#fft test2

fs=1/365
freq = np.fft.fftfreq(fshift.shape[-1], fs)
plt.plot(np.fft.fftshift(freq),fshift)
```
