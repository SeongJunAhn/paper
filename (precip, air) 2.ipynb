```python
new_lat=(precip_lat_data+new_air_lat_data)/2
new_lon=(precip_lon_data+air_lon_data)/2
```

```python
print(new_lat),print(new_lon)
```

```python
labels=[]
for i in range(len(precip_kmeans)):
    line=[]
    line.append(precip_kmeans[i])
    line.append(air_kmeans[i])
    labels.append(line)
```

```python
for i in range(len(labels)):
    print(labels[i])
```

```python
for i in range(5):
    for j in range(5):
        print([i,j],[i,j] in labels,5*i+j)
```

```python
new_labels=[]
for i in range(len(labels)):
    for j in range(5):
        for k in range(5):
            if labels[i]==[j,k]:
                if j==0:
                    new_labels.append(k)
                elif k==0:
                    new_labels.append(5*j)
                else:
                    new_labels.append((5*j+k))
```

```python
#144 
test=[]
for i in range(72):
    test.append(new_labels[144*i:144*(i+1)])
```

```python
new_test=[]#
for i in range(72):
    front=test[i][:72]
    back=test[i][72:]
    new_test.append(back+front)
```

```python
for i in range(72):
    print(new_test[i][:72]==test[i][72:],new_test[i][72:]==test[i][:72])
```

```python
#
flatten=np.array(new_test).flatten().tolist()
```

```python
import os
os.environ['PROJ_LIB'] = r'C:\Users\yunju\Anaconda3\pkgs\proj4-5.2.0-ha925a31_1\Library\share'
from mpl_toolkits.basemap import Basemap
import matplotlib.pyplot as plt
from numpy import meshgrid
map = Basemap()

plt.figure(figsize=(22,10))
map.drawcoastlines()
map.drawmapboundary()
map.drawparallels(np.arange(-90.,90.,30.),labels=[1,0,0,0])
map.drawmeridians(np.arange(-180.,180.,60.),labels=[0,0,0,1])

x=[]
for i in range(len(new_lon)):
    if new_lon[i]>180:
        x.append(new_lon[i]-360)
    else:
        x.append(new_lon[i])
x.sort()
y=new_lat
data=[]
for i in range(72):
    data.append(flatten[144*i:144*(i+1)])



xx,yy=meshgrid(x,y)

cmap = plt.get_cmap('Pastel2')#

colormesh = map.pcolormesh(xx, yy, data, cmap=cmap)
contour = map.contourf(xx, yy, data, range(500, 1350, 1), colors = 'k', linestyles = 'dotted')


cb = map.colorbar(colormesh, location='bottom', label="contour lines")


plt.title('Clustering by precipitation and temperature',fontsize=25)                
plt.show()
```
