deleting empty kmeans labels

```python
labels=[]
for i in range(len(precip_kmeans)):
    line=[]
    line.append(precip_kmeans[i])
    line.append(air_kmeans[i])
    labels.append(line)
```

```python
tf=[]
for i in range(5):
    for j in range(5):
        tf.append([i,j] in labels)
        print([i,j],[i,j] in labels,5*i+j)
```

```python
target=[]
for i in range(len(tf)):
    if tf[i]==True:
        target.append(i)
    if tf[i]==False:
        target.append(None)
```

```python
new_labels=[]
for i in range(len(labels)):
    for j in range(5):
        for k in range(5):
            a=labels[i]
            if a==[j,k]:
                if j==0:                    
                    new_labels.append(k-target[:5*j+k].count(None))
                elif k==0:
                    new_labels.append(5*j-target[:5*j+k].count(None))
                else:
                    new_labels.append(5*j+k-target[:5*j+k].count(None))                          
```

labels relocation for basemap plotting

```python
#144개씩 끊어서
test=[]
for i in range(72):
    test.append(new_labels[144*i:144*(i+1)])

#쉬프트
new_test=[]
for i in range(72):
    front=test[i][:72]
    back=test[i][72:]
    new_test.append(back+front)

#재결합
flatten=np.array(new_test).flatten().tolist()
```

visualizing

```python
import os
os.environ['PROJ_LIB'] = r'C:\Users\yunju\Anaconda3\pkgs\proj4-5.2.0-ha925a31_1\Library\share'
from mpl_toolkits.basemap import Basemap
import matplotlib.pyplot as plt
from numpy import meshgrid
map = Basemap()

plt.figure(figsize=(22,10))
map.drawcoastlines()
map.drawmapboundary()
map.drawparallels(np.arange(-90.,90.,30.),labels=[1,0,0,0])
map.drawmeridians(np.arange(-180.,180.,60.),labels=[0,0,0,1])

x=[]
for i in range(len(new_lon)):
    if new_lon[i]>180:
        x.append(new_lon[i]-360)
    else:
        x.append(new_lon[i])
x.sort()
y=new_lat
data=[]
for i in range(72):
    data.append(flatten[144*i:144*(i+1)])



xx,yy=meshgrid(x,y)

cmap = plt.get_cmap('tab20')

colormesh = map.pcolormesh(xx, yy, data, cmap=cmap)
contour = map.contourf(xx, yy, data, range(500, 1350, 50), colors = 'k')

cb = map.colorbar(colormesh, location='bottom', label="contour lines")
cb.set_ticks([i for i in range(20)])

plt.title('Clustering by precipitation and temperature',fontsize=25)                
plt.show()
```
